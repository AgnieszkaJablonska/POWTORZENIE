```python
# Import the needed fuctions
from matplotlib import pyplot as plt 
#matplotlib.pyplot is the standard for making plots and is derived from Matlab
import numpy as np
```

### Lets plot our data from last class!!!  YAY WE GET TO FINALLY MAKE PLOTS!

```python
# READ IN DATA FILE/
data = np.genfromtxt("munich_temperatures_average_with_bad_data.txt")
Date = data[:,0]
Temp = data[:,1] 

# PLOT TO VERIFY FILE READ 
# This creates a figure object and sets the size
fig1 = plt.figure(figsize=(20,10))
# This allows subplots, for now this is set to 1 plot
ax1 = fig1.add_subplot(111)

ax1.scatter(Date,Temp,s=15,c='k',alpha=0.5)

# This increases the 
ax1.xaxis.set_tick_params(labelsize=20)
ax1.yaxis.set_tick_params(labelsize=20)

```

## A second plot

You will first need to download the plain text data file "CG20.dat", that we will read in to have data to plot.

```python
# This will read in the data into a numpy array
data = np.genfromtxt("CG20.csv", delimiter=",")
# This will seperate specific columns of the data into newly created variables
AgeNN = data[:,14]  # Log Age 
DistPc = data[:,17] # Distance
XGC = data[:,18]   # X-coordinate
YGC = data[:,19]   # Y-coordinate
ZGC = data[:,20]   # Z-coordinate
RGC = data[:,21]   # Radius-coordinate
```

__Next will will gerenerate a first plot of X vs Y position.__

This will introduce the first options you will have.

```python
# Creating a scatter plot in log log space using an object oriented approach
fig = plt.figure()                           # makes a figure
ax = fig.add_subplot(111)                    # make a 1,1,1 (single) figure
ax.scatter(XGC,YGC,s=5,c='k',alpha=0.3)      # makes a scatter plot (s = size, c = color, alpha = transparency)
```

```python
# Changing the size of the plot -- figsize
fig = plt.figure(figsize=(5,5))          # makes a figure (allow us to change the size of the figure)
ax = fig.add_subplot(111)                  # make a 1,1,1 (single) figure
ax.scatter(XGC,YGC,s=5,c='k',alpha=0.3)    # makes a scatter plot (s = size, c = color, alpha = transparency)

```

Well this looks like a nice plot, but if you gave this to someone could they tell what it was a plot of?

We need add labels to inform the reader, also we will limit the plot to only show a portion of the data.

```python
# Reread the data to get ALL of the columns
data = np.genfromtxt("CG20.csv", delimiter=",", names = True)

#the first line of files containing names of columns will be used), thanks to line names = True

print(data.dtype.names)
# this is the same as above
fig = plt.figure(figsize=(20,15))
ax = fig.add_subplot(111)
ax.scatter(data['X'],data['Y'],s=20,c='k',alpha=0.3)    # makes a scatter plot (s = size, c = color, alpha = transparency)


# Now, we will label the axis and set the limits of the axis and increase ticksize of the axis
# Set the start and end of range you want to display on the x and yaxis 
ax.set_xlim(-2000,2000)
ax.set_ylim(-1500,1500)

# add the labels
ax.set_xlabel("$X_{GC}$ $(pc)$",size=60)   # allows LaTeX style formating
ax.set_ylabel("$Y_{GC}$ $(pc)$",size=60)   # allows LaTeX style formating

# increase the size of the numbers on the axes
ax.xaxis.set_tick_params(labelsize=50)
ax.yaxis.set_tick_params(labelsize=50)
```

```python
# this is the same as above
fig = plt.figure(figsize=(20,15))
ax = fig.add_subplot(111)

# Sizing the points according to a 3rd attribute - rvir
ax.scatter(data['X'],data['Y'],s=200*data['r50'],c='k',alpha=0.3)  

#r50 - how big is this on the sky, alpha - says how transparent my points are, when 1 is the most opaque ( not transparent)

# Now, we will label the axis and set the limits of the axis and increase ticksize of the axis
# Set the start and end of range you want to display on the x and yaxis 
ax.set_xlim(-2000,2000)
ax.set_ylim(-1500,1500)

# add the labels
ax.set_xlabel("$X_{GC}$ $(pc)$",size=20)   # allows LaTeX style formating
ax.set_ylabel("$Y_{GC}$ $(pc)$",size=20)   # allows LaTeX style formating

# increase the size of the numbers on the axes
ax.xaxis.set_tick_params(labelsize=20)
ax.yaxis.set_tick_params(labelsize=20)
```

```python
# this is the same as above
fig = plt.figure(figsize=(20,15))
ax = fig.add_subplot(111)

# Coloring the points by a 4th attribute and adding a color bar
image = ax.scatter(data['X'],data['Y'],s=200*data['r50'],c=data['AgeNN'],edgecolor='m',cmap="plasma",alpha=0.5)
bar = fig.colorbar(image,orientation="vertical",pad=0.01)
bar.set_label('$log_{10} (Age)$',size=15)


# python documentation, numpy documentation
#colors: c, m, y, k, r, g, b
#: cyan. magenta, yellow, black, red, green, blue

# in google py.plot on this site i can find the commands of mathlab
# cmap: rainbow, viridis, plasma, RdPu, inferno, magma, cool, spring, Pastel1, flag, prism,

# Now, we will label the axis and set the limits of the axis and increase ticksize of the axis
# Set the start and end of range you want to display on the x and yaxis 
ax.set_xlim(-2000,2000)
ax.set_ylim(-1500,1500)

# add the labels
ax.set_xlabel("$X_{GC}$ $(pc)$",size=20)   # allows LaTeX style formating
ax.set_ylabel("$Y_{GC}$ $(pc)$",size=20)   # allows LaTeX style formating

# increase the size of the numbers on the axes
ax.xaxis.set_tick_params(labelsize=20)
ax.yaxis.set_tick_params(labelsize=20)


```

Now a plot with RAdeg vs. DEdeg with size scaled by r50 and color coded by the log10 of nbstars07.  Note make it as readable as possible!

```python
# flie CG20
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)

# Coloring the points by a 4th attribute and adding a color bar
image = ax.scatter(data['RAdeg'],data['DEdeg'],s=200*data['r50'],c = np.log(data['nbstars07']),edgecolor='m',cmap="plasma",alpha=0.5)
bar = fig.colorbar(image,orientation="vertical",pad=0.01)
bar.set_label('$log_{10} $(Number of stars)',size=15)

ax.set_xlabel("RA $(deg)$",size=20)   # allows LaTeX style formating
ax.set_ylabel("DE $(deg)$",size=20)   # allows LaTeX style formating

```

```python
# Adding multiple plots
fig = plt.figure(figsize=(10,8))

fig.subplots_adjust(hspace=0.50, wspace=0.50)

# 221 - 2x2 plot 1, so it means we have 4 plots, placed like letter z, plot 1 (top left, plot 2 on the right of plot 1)
ax1 = fig.add_subplot(221)
image1 = ax1.scatter(data['GLON'],data['GLAT'], c = data['AgeNN'], s = 200*data['r50'], alpha = 0.5)

bar = fig.colorbar(image1,orientation="vertical",pad=0.01)
bar.set_label('$log_{10} $(years)',size=15)

ax1.set_xlabel("GLON $(deg)$",size=20)   # allows LaTeX style formating
ax1.set_ylabel("GLAT $(deg)$",size=20)   # allows LaTeX style formating


# ax2 = fig.add_subplot(222)
# ax2.scatter(data['pmRA'],data['pmDE'])

ax2 = fig.add_subplot(222)
image2 = ax2.scatter(data['pmRA'],data['pmDE'], c = data['AgeNN'], s = 200*data['r50'], alpha = 0.5)

bar = fig.colorbar(image2,orientation="vertical",pad=0.01)
bar.set_label('$log_{10} $(years)',size=15)

ax2.set_xlabel("pmRA $(deg)$",size=20)   # allows LaTeX style formating
ax2.set_ylabel("pmDE $(deg)$",size=20)   # allows LaTeX style formating


# ax3 = fig.add_subplot(223)
# ax3.scatter(data['Rgc'],data['Z'])

ax3 = fig.add_subplot(223)
image3 = ax3.scatter(data['Rgc'],data['Z'], c = data['AgeNN'], s = 200*data['r50'], alpha = 0.5)

bar = fig.colorbar(image3,orientation="vertical",pad=0.01)
bar.set_label('$log_{10} $(years)',size=15)

ax3.set_xlabel("Rgc $(Pc)$",size=20)   # allows LaTeX style formating
ax3.set_ylabel("Z $(Pc)$",size=20)   # allows LaTeX style formating

# ax4 = fig.add_subplot(224)
# ax4.scatter(data['AgeNN'],data['DistPc'])

ax4 = fig.add_subplot(224)
image4 = ax4.scatter(data['AgeNN'],data['DistPc'], c = data['AgeNN'], s = 200*data['r50'], alpha = 0.5)

bar = fig.colorbar(image4,orientation="vertical",pad=0.01)
bar.set_label('$log_{10} $(years)',size=15)

ax4.set_xlabel("AgeNN $(years)$",size=20)   # allows LaTeX style formating
ax4.set_ylabel("Dist $(Pc)$",size=20)   # allows LaTeX style formating
```
